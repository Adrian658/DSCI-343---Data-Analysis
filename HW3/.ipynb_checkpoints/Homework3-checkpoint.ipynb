```python
import pandas as pd
import numpy as np
from scipy.linalg import lstsq
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import statistics 
from sklearn.metrics import f1_score
```

```python
p1_training = pd.read_csv("p1_features_training.csv", header=None)
p1_training_out = pd.read_csv("p1_outcome_training.csv", header=None)
p1_test = pd.read_csv("p1_features_test.csv", header=None)
p1_test_out = pd.read_csv("p1_outcome_test.csv", header=None)
```

```python
p1_training
```

## Problem 1

```python
def linear_model(feature_vals, weights):
    output = weights[0]
    for weight, val in zip(weights[1:], feature_vals):
        output += weight*val
    return output
```

```python
def find_weights(data, outputs, col_indices):
    constant_vec = np.array([1]*len(outputs))
    feature_df = p1_training[col_indices].values
    feature_df = np.append(constant_vec[:, np.newaxis], feature_df, 1)
    model = lstsq(feature_df, outputs)
    return model[0]
```

```python
def run_model(data, weights):
    test_vec= []
    for feature in data.values:
        res = linear_model(feature, weights)
        test_vec.append(res)
    return test_vec
```

```python
def visualize_outputs(true, predicted):
    x = range(0, len(true))
    print("R^2 score:", r2_score(true, predicted))
    plt.plot(x, true, label="True Output")
    plt.plot(x, predicted, label="Predicted Output")
    plt.legend()
    plt.show()
```

```python
def complete_run_model(num_features):
    train_data = p1_training
    train_data_out = p1_training_out
    test_data = p1_test
    test_data_out = p1_test_out
    
    print("Training and running model on training data")
    weights = find_weights(train_data, train_data_out[0], range(num_features))
    training_predictions = run_model(train_data, weights)
    visualize_outputs(train_data_out[0], training_predictions)
    
    print("Running model on test data")
    test_predictions = run_model(test_data, weights)
    visualize_outputs(test_data_out[0], test_predictions)
```

```python
def plot_r2_scores():
    r_train = [0.0, 0.2646, 0.3607, 0.3709, 0.3744, 0.3771, 0.4042, 0.4214, 0.4214, 0.4218, 0.4335, 'training']
    r_test = [-0.0231, 0.1979, 0.2000, 0.2302, 0.2397, 0.2404, 0.1492, 0.1017, 0.1018, 0.0990, 0.06022, 'test']
    frame = pd.DataFrame([r_train, r_test])
    frame.set_index(11, inplace=True)
    x = range(0, len(r_train)-1)
    plt.plot(x, frame.loc['training'], label='Training R-score')
    plt.plot(x, frame.loc['test'], label='Test R-score')
    plt.legend()
    plt.show()
```

```python
complete_run_model(5)
```

```python
plot_r2_scores()
```

## Problem 2

```python
votes = pd.read_csv("p2_congress_1984_votes.csv", header=None)
affiliations = pd.read_csv("p2_congress_1984_party_affiliations.csv", header=None)
```

```python
votes
```

```python
affiliations
```

```python
pca = PCA(n_components=2)
components = pca.fit_transform(votes.values)
```

```python
pca.explained_variance_ratio_
```

```python
r = pd.DataFrame(pca.inverse_transform(components))
```

```python
variance_percentages = pca.explained_variance_ratio_ * 100
variance_percentages_cumu = [0]
for percent in variance_percentages:
    try:
        temp = percent + variance_percentages_cumu[-1]
    except:
        temp = percent
    variance_percentages_cumu.append(temp)
```

```python
variance_percentages
```

```python
plt.plot(range(17), variance_percentages_cumu)
plt.xlabel("Primary Components")
plt.ylabel("Cumulative Variance")
plt.show()
```

```python
principalDf = pd.DataFrame(data = components, columns = ['PC 1', 'PC 2', 'PC 3', 'PC 4', 'PC 5', 'PC 6', 'PC 7', 'PC 8', 'PC 9', 'PC 10', 'PC 11', 'PC 12', 'PC 13', 'PC 14', 'PC 15', 'PC 16'])
```

```python
finalDf = pd.concat([principalDf, affiliations[[0]]], axis = 1)
```

```python
finalDf["const"] = range(len(finalDf))
```

```python
one_two = ['PC 1', 'PC 2']
one_three = ['PC 1', 'PC 3']
two_three = ['PC 2', 'PC 3']
for PC in [one_two, one_three, two_three]: 
    fig = plt.figure(figsize = (8,8))
    ax = fig.add_subplot(1,1,1) 
    ax.set_xlabel(PC[0], fontsize = 15)
    ax.set_ylabel(PC[1], fontsize = 15)
    targets = ['Democrat', 'Republican']
    colors = ['r', 'g']
    for target, color in zip(targets,colors):
        indicesToKeep = finalDf[0] == target
        ax.scatter(finalDf.loc[indicesToKeep, PC[0]]
                   , finalDf.loc[indicesToKeep, PC[1]]
                   , c = color
                   , s = 50)
    ax.legend(targets)
    ax.grid()
```

```python
k_means = KMeans(n_clusters=2).fit(components)
labels = k_means.labels_
new_labels = ['Democrat' if label == 0 else 'Republican' for label in labels]
df = pd.DataFrame(new_labels)
df
```

```python
f1_score(affiliations.values, df.values, average='micro')
```
