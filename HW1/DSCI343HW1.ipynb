### Generate alpha parameter of power law distribution

```python
def power_law(target):

    n = len(target)
    minim = target.min()[1]
    sumtotal = 0
    for row in target.values:
        divide = row[1] / minim
        divide = np.log(divide)
        sumtotal+=divide

    alpha = 1 + n * (sumtotal)**-1
    
    return alpha
```

### Generate lambda parameter of exponential distribution

```python
def exponential(target):
    n = len(target)
    sumtotal = 0
    for row in target.values:
        sumtotal += row[1]

    lambda_ = n / sumtotal
    return lambda_
```

### Generate a and b parameters of uniform distribution

```python
def uniform(target):

    a = target.min()[1]
    b = target.max()[1]
    
    return [a, b]
```

### Generate mew and sigma parameters of normal distribution

```python
def normal(target):
    
    n = len(target)
    sumtotal = 0
    for row in target.values:
        sumtotal += row[1]

    mew = sumtotal / n

    variance_accumulator = 0
    for row in target.values:
        diff = row[1] - mew
        variance_accumulator += diff**2

    sigma_squared = variance_accumulator / n
    sigma = sigma_squared**.5

    return [mew, sigma]
```
